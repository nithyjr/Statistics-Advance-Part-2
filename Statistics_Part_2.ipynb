1. What is hypothesis testing in statistics?

Answer: Hypothesis testing is a statistical method used to make inferences about a population based on sample data. It involves formulating two competing hypotheses (null and alternative) and determining whether the data provides sufficient evidence to reject the null hypothesis.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer:
Null Hypothesis (H₀): Suggests that there is no effect, difference, or relationship in the population (the status quo).
Alternative Hypothesis (H₁ or Ha): Suggests that there is an effect, difference, or relationship (what you're trying to prove).

3. What is the significance level in hypothesis testing, and why is it important?

Answer: The significance level (α) is the probability of rejecting the null hypothesis when it is actually true (Type 1 error). It determines the threshold for statistical significance and is typically set at 0.05 (5%).

4. What does a P-value represent in hypothesis testing?

Answer: The P-value is the probability of obtaining a result at least as extreme as the one observed, assuming the null hypothesis is true. A smaller P-value suggests stronger evidence against the null hypothesis.

5. How do you interpret the P-value in hypothesis testing?

Answer:
If the P-value ≤ α (significance level), reject the null hypothesis.
If the P-value > α, fail to reject the null hypothesis.

6. What are Type 1 and Type 2 errors in hypothesis testing?

Answer:
Type 1 Error (False Positive): Rejecting the null hypothesis when it is actually true.
Type 2 Error (False Negative): Failing to reject the null hypothesis when the alternative hypothesis is true.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

Answer:
One-tailed test: Tests for a deviation in one direction (either greater than or less than).
Two-tailed test: Tests for deviations in both directions (either greater than or less than).

8. What is the Z-test, and when is it used in hypothesis testing?

Answer: The Z-test is used when the sample size is large (n ≥ 30) and the population standard deviation is known. It compares sample means or proportions with population means or proportions to determine if a significant difference exists.

10. What is the T-distribution, and when should it be used instead of the normal distribution?

Answer: The T-distribution is used when the sample size is small (n < 30) and/or the population standard deviation is unknown. It has heavier tails to account for more variability in small samples.

11. What is the difference between a Z-test and a T-test?

Answer:
Z-test: Used when the population standard deviation is known and the sample size is large.
T-test: Used when the population standard deviation is unknown and the sample size is small.

12. What is the T-test, and how is it used in hypothesis testing?

Answer: The T-test is used to determine if there is a significant difference between the means of two groups, especially when the sample size is small and the population standard deviation is unknown.

13. What is the relationship between Z-test and T-test in hypothesis testing?

Answer: Both tests compare sample means with population means. The Z-test is used for large samples or when the population standard deviation is known, while the T-test is used for small samples or when the population standard deviation is unknown.

14. What is a confidence interval, and how is it used to interpret statistical results?

Answer: A confidence interval is a range of values used to estimate a population parameter, such as a mean. It provides an interval in which we are confident the true value lies, given a specific confidence level (e.g., 95%).

15. What is the margin of error, and how does it affect the confidence interval?

Answer: The margin of error represents the potential variability in the sample estimate. It affects the width of the confidence interval: a larger margin of error results in a wider interval, indicating more uncertainty about the true population parameter.

16. How is Bayes' Theorem used in statistics, and what is its significance?

Answer: Bayes' Theorem is used to update the probability of a hypothesis based on new evidence. It is important in Bayesian statistics, which involves updating prior knowledge with data to refine parameter estimates.

17. What is the Chi-square distribution, and when is it used?

Answer: The Chi-square distribution is used for hypothesis testing in categorical data. It is commonly used in tests of independence and goodness-of-fit tests.

18. What is the Chi-square goodness of fit test, and how is it applied?

Answer: This test compares the observed frequencies to the expected frequencies under the null hypothesis to determine whether the sample data fits a specific distribution.

19. What is the F-distribution, and when is it used in hypothesis testing?

Answer: The F-distribution is used to compare variances of two or more groups. It is frequently used in Analysis of Variance (ANOVA) and to compare the variances in hypothesis testing.

20. What is an ANOVA test, and what are its assumptions?

Answer: The Analysis of Variance (ANOVA) test compares the means of three or more groups to see if there are significant differences. Assumptions include normality of data, homogeneity of variance, and independence of observations.

21. What are the different types of ANOVA tests?

Answer:
One-way ANOVA: Compares means of three or more groups based on one independent variable.
Two-way ANOVA: Compares means of groups based on two independent variables, considering both main effects and interaction effects.

22. What is the F-test, and how does it relate to hypothesis testing?

Answer: The F-test compares the variances of two or more groups to see if they are significantly different. It is used in ANOVA to test if all group means are equal.


```python
#  1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
# interpret the results

import numpy as np
import scipy.stats as stats

# Function to perform Z-test
def z_test(sample_mean, population_mean, sample_std, sample_size, alpha=0.05):
    # Calculate Z-statistic
    z_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))

    # Calculate p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    # Print results
    print(f"Z-statistic: {z_statistic}")
    print(f"P-value: {p_value}")

    # Interpret the result
    if p_value < alpha:
        print(f"Reject the null hypothesis. There is enough evidence to suggest the sample mean is different from the population mean.")
    else:
        print(f"Fail to reject the null hypothesis. There is not enough evidence to suggest the sample mean is different from the population mean.")

# Example usage
sample_mean = 105      # Mean of the sample
population_mean = 100  # Known population mean
sample_std = 15        # Standard deviation of the sample
sample_size = 30       # Size of the sample
alpha = 0.05           # Significance level

# Perform the Z-test
z_test(sample_mean, population_mean, sample_std, sample_size, alpha)
```

```python
# 3.  Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
import numpy as np
import scipy.stats as stats

# Function to perform a one-sample Z-test
def one_sample_z_test(sample_data, population_mean, alpha=0.05):
    # Step 1: Calculate sample statistics
    sample_size = len(sample_data)
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation

    # Step 2: Calculate the Z-statistic
    standard_error = sample_std / np.sqrt(sample_size)
    z_statistic = (sample_mean - population_mean) / standard_error

    # Step 3: Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    # Step 4: Output results
    print(f"Sample Mean: {sample_mean}")
    print(f"Z-statistic: {z_statistic}")
    print(f"P-value: {p_value}")

    # Step 5: Hypothesis Testing
    if p_value < alpha:
        print(f"Reject the null hypothesis. There is enough evidence to suggest the sample mean is different from the population mean.")
    else:
        print(f"Fail to reject the null hypothesis. There is not enough evidence to suggest the sample mean is different from the population mean.")

# Example Usage
np.random.seed(42)  # For reproducibility

# Simulating sample data
population_mean = 100    # Known population mean
sample_data = np.random.normal(loc=population_mean, scale=15, size=30)  # Sample of 30 elements

# Perform the one-sample Z-test
one_sample_z_test(sample_data, population_mean, alpha=0.05)
```

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Function to perform a two-tailed Z-test
def two_tailed_z_test(sample_data, population_mean, alpha=0.05):
    # Step 1: Calculate sample statistics
    sample_size = len(sample_data)
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation

    # Step 2: Calculate the Z-statistic
    standard_error = sample_std / np.sqrt(sample_size)
    z_statistic = (sample_mean - population_mean) / standard_error

    # Step 3: Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    # Step 4: Output results
    print(f"Sample Mean: {sample_mean}")
    print(f"Z-statistic: {z_statistic}")
    print(f"P-value: {p_value}")

    # Step 5: Hypothesis Testing
    if p_value < alpha:
        print(f"Reject the null hypothesis. There is enough evidence to suggest the sample mean is different from the population mean.")
    else:
        print(f"Fail to reject the null hypothesis. There is not enough evidence to suggest the sample mean is different from the population mean.")

    # Plotting the Z-distribution and decision region
    plot_z_distribution(z_statistic, alpha)

# Function to plot the Z-distribution and decision region
def plot_z_distribution(z_statistic, alpha):
    # Create a range of z values for plotting the standard normal distribution
    z_values = np.linspace(-4, 4, 1000)
    z_pdf = stats.norm.pdf(z_values)  # PDF of the standard normal distribution

    # Plot the normal distribution curve
    plt.plot(z_values, z_pdf, label="Standard Normal Distribution", color='blue')

    # Critical z-values for a two-tailed test with significance level alpha
    z_critical = stats.norm.ppf(1 - alpha / 2)

    # Shade the rejection regions (critical regions)
    plt.fill_between(z_values, 0, z_pdf, where=(z_values < -z_critical) | (z_values > z_critical),
                     color='red', alpha=0.5, label='Rejection Region')

    # Highlight the calculated Z-statistic on the plot
    plt.axvline(z_statistic, color='green', linestyle='--', label=f'Z-statistic = {z_statistic:.2f}')

    # Labels and title
    plt.title('Two-Tailed Z-Test Decision Region')
    plt.xlabel('Z-Score')
    plt.ylabel('Probability Density')

    # Add a legend
    plt.legend(loc='upper left')

    # Show the plot
    plt.grid(True)
    plt.show()

# Example Usage
np.random.seed(42)  # For reproducibility

# Simulating sample data
population_mean = 100    # Known population mean
sample_data = np.random.normal(loc=population_mean, scale=15, size=30)  # Sample of 30 elements

# Perform the two-tailed Z-test
two_tailed_z_test(sample_data, population_mean, alpha=0.05)
```

```python
# 5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Function to visualize Type 1 and Type 2 Errors
def visualize_errors(population_mean, sample_size, population_std, alpha=0.05, true_mean=105):
    # Critical Z-values for the given alpha level (two-tailed test)
    z_critical = stats.norm.ppf(1 - alpha / 2)

    # Visualize the normal distributions for H0 (null hypothesis) and H1 (alternative hypothesis)
    # Under the null hypothesis (H0), we assume the mean is population_mean
    # Under the alternative hypothesis (H1), we assume the true mean is different (true_mean)

    # Generate range of Z values for plotting
    z_values = np.linspace(-4, 4, 1000)
    z_pdf_null = stats.norm.pdf(z_values, loc=0, scale=1)  # Under H0 (Standard normal)
    z_pdf_alt = stats.norm.pdf(z_values, loc=(true_mean - population_mean) / (population_std / np.sqrt(sample_size)), scale=1)  # Under H1

    # Plot the normal distribution curves for both H0 and H1
    plt.plot(z_values, z_pdf_null, label='Null Hypothesis (H0)', color='blue')
    plt.plot(z_values, z_pdf_alt, label='Alternative Hypothesis (H1)', color='orange')

    # Shade the rejection region (alpha) under H0
    plt.fill_between(z_values, 0, z_pdf_null, where=(z_values < -z_critical) | (z_values > z_critical),
                     color='red', alpha=0.5, label='Rejection Region (Type 1 Error Area)')

    # Shade the Type 2 error region under H1 (fail to reject H0 when H1 is true)
    plt.fill_between(z_values, 0, z_pdf_alt, where=(z_values >= -z_critical) & (z_values <= z_critical),
                     color='green', alpha=0.5, label='Type 2 Error Area')

    # Mark the critical values (Z-critical) on the plot
    plt.axvline(-z_critical, color='black', linestyle='--', label=f'Critical Z = {-z_critical:.2f}')
    plt.axvline(z_critical, color='black', linestyle='--', label=f'Critical Z = {z_critical:.2f}')

    # Labels and title
    plt.title(f'Type 1 and Type 2 Errors Visualization\nAlpha = {alpha}, True Mean = {true_mean}')
    plt.xlabel('Z-Score')
    plt.ylabel('Probability Density')

    # Add a legend
    plt.legend(loc='upper left')

    # Show the plot
    plt.grid(True)
    plt.show()

# Example usage:
population_mean = 100  # Known population mean
population_std = 15    # Known population standard deviation
sample_size = 30       # Sample size
alpha = 0.05           # Significance level
true_mean = 105        # True mean (alternative hypothesis)

# Visualize Type 1 and Type 2 errors
visualize_errors(population_mean, sample_size, population_std, alpha, true_mean)
```

```python
# 6.Write a Python program to perform an independent T-test and interpret the results.
import numpy as np
from scipy import stats

# Function to perform independent T-test
def independent_t_test(group1, group2, alpha=0.05):
    # Perform the T-test
    t_stat, p_value = stats.ttest_ind(group1, group2)

    # Print the results
    print(f"T-statistic: {t_stat}")
    print(f"P-value: {p_value}")

    # Interpret the result
    if p_value < alpha:
        print(f"Reject the null hypothesis. There is enough evidence to suggest the means are different.")
    else:
        print(f"Fail to reject the null hypothesis. There is not enough evidence to suggest the means are different.")

# Example usage
np.random.seed(42)  # For reproducibility

# Simulate two independent groups with normal distribution
# Group 1: Mean = 100, Standard deviation = 10, Size = 30
group1 = np.random.normal(loc=100, scale=10, size=30)

# Group 2: Mean = 105, Standard deviation = 12, Size = 30
group2 = np.random.normal(loc=105, scale=12, size=30)

# Perform the independent T-test
independent_t_test(group1, group2, alpha=0.05)
```

```python
# 7. Perform a paired sample T-test using Python and visualize the comparison results
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Function to perform a paired sample T-test
def paired_t_test(before, after, alpha=0.05):
    # Perform the paired T-test
    t_stat, p_value = stats.ttest_rel(before, after)

    # Print the results
    print(f"T-statistic: {t_stat}")
    print(f"P-value: {p_value}")

    # Interpret the result
    if p_value < alpha:
        print(f"Reject the null hypothesis. There is enough evidence to suggest a significant difference.")
    else:
        print(f"Fail to reject the null hypothesis. There is not enough evidence to suggest a significant difference.")

    # Visualize the comparison results
    visualize_comparison(before, after)

# Function to visualize the comparison (before vs after)
def visualize_comparison(before, after):
    # Calculate the differences between the paired samples
    differences = after - before

    # Plotting the data
    plt.figure(figsize=(10, 6))

    # Scatter plot for before vs after (paired samples)
    plt.scatter(range(len(before)), before, color='blue', label='Before', alpha=0.7)
    plt.scatter(range(len(after)), after, color='orange', label='After', alpha=0.7)

    # Plot the differences as a bar plot
    plt.plot(range(len(differences)), differences, label='Differences (After - Before)', color='green', marker='o')

    # Labels and title
    plt.title('Paired Sample Comparison (Before vs After)')
    plt.xlabel('Sample Index')
    plt.ylabel('Values')

    # Add a legend
    plt.legend(loc='best')

    # Show the plot
    plt.grid(True)
    plt.show()

# Example usage
np.random.seed(42)  # For reproducibility

# Simulating paired data (Before and After treatment)
before_treatment = np.random.normal(loc=100, scale=10, size=30)  # Before treatment values
after_treatment = before_treatment + np.random.normal(loc=5, scale=5, size=30)  # After treatment values (simulating improvement)

# Perform the paired sample T-test
paired_t_test(before_treatment, after_treatment, alpha=0.05)
```

```python
# 8.Simulate data and perform both Z-test and T-test, then compare the results using Python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Function to perform Z-test
def z_test(sample_data, population_mean, population_std, alpha=0.05):
    sample_size = len(sample_data)
    sample_mean = np.mean(sample_data)
    standard_error = population_std / np.sqrt(sample_size)

    # Z-statistic
    z_statistic = (sample_mean - population_mean) / standard_error

    # p-value for two-tailed Z-test
    p_value_z = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    print(f"Z-test: Z-statistic = {z_statistic}, p-value = {p_value_z}")
    return p_value_z

# Function to perform T-test
def t_test(sample_data, population_mean, alpha=0.05):
    sample_size = len(sample_data)
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation

    # T-statistic
    t_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))

    # p-value for two-tailed T-test
    p_value_t = 2 * (1 - stats.t.cdf(abs(t_statistic), df=sample_size - 1))

    print(f"T-test: T-statistic = {t_statistic}, p-value = {p_value_t}")
    return p_value_t

# Simulate data
np.random.seed(42)  # For reproducibility

# Assume population mean = 100 and population standard deviation = 15
population_mean = 100
population_std = 15

# Simulate a sample of 30 data points (sample size = 30)
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=30)

# Perform both Z-test and T-test
alpha = 0.05
p_value_z = z_test(sample_data, population_mean, population_std, alpha)
p_value_t = t_test(sample_data, population_mean, alpha)

# Compare results
if p_value_z < alpha:
    print("Z-test: Reject the null hypothesis (mean is significantly different).")
else:
    print("Z-test: Fail to reject the null hypothesis (no significant difference).")

if p_value_t < alpha:
    print("T-test: Reject the null hypothesis (mean is significantly different).")
else:
    print("T-test: Fail to reject the null hypothesis (no significant difference).")
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
import scipy.stats as stats

# Function to calculate the confidence interval for a sample mean
def confidence_interval(sample_data, confidence_level=0.95):
    # Sample size (n), sample mean, and sample standard deviation (s)
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Using ddof=1 for sample standard deviation

    # t-critical value for the given confidence level
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df=n-1)

    # Standard error
    standard_error = sample_std / np.sqrt(n)

    # Confidence Interval calculation
    margin_of_error = t_critical * standard_error
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage:
np.random.seed(42)  # For reproducibility

# Simulate a sample with 30 data points
sample_data = np.random.normal(loc=100, scale=15, size=30)

# Calculate the 95% confidence interval for the sample mean
lower_bound, upper_bound = confidence_interval(sample_data, confidence_level=0.95)

print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
# 10.Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
import scipy.stats as stats

# Function to calculate the margin of error for a given confidence level
def margin_of_error(sample_data, confidence_level=0.95):
    # Sample size (n), sample mean, and sample standard deviation (s)
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Using ddof=1 for sample standard deviation

    # t-critical value for the given confidence level
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df=n-1)

    # Standard error
    standard_error = sample_std / np.sqrt(n)

    # Margin of error calculation
    margin_of_error = t_critical * standard_error

    return margin_of_error

# Example usage:
np.random.seed(42)  # For reproducibility

# Simulate a sample with 30 data points
sample_data = np.random.normal(loc=100, scale=15, size=30)

# Calculate the margin of error for the 95% confidence level
moe = margin_of_error(sample_data, confidence_level=0.95)

print(f"Margin of Error: {moe:.2f}")
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayes_theorem(prior_H, likelihood_D_given_H, likelihood_D_given_not_H, prior_not_H):
    """
    Calculate the posterior probability using Bayes' Theorem.

    Args:
    - prior_H (P(H)): The prior probability of having the disease.
    - likelihood_D_given_H (P(D|H)): The likelihood of the test being positive given the disease.
    - likelihood_D_given_not_H (P(D|¬H)): The likelihood of the test being positive without the disease.
    - prior_not_H (P(¬H)): The prior probability of not having the disease.

    Returns:
    - posterior_H_given_D (P(H|D)): The posterior probability of having the disease given a positive test result.
    """

    # Evidence (P(D)) is the total probability of a positive test result
    evidence_D = (likelihood_D_given_H * prior_H) + (likelihood_D_given_not_H * prior_not_H)

    # Bayes' Theorem: Posterior = (Likelihood * Prior) / Evidence
    posterior_H_given_D = (likelihood_D_given_H * prior_H) / evidence_D

    return posterior_H_given_D

# Example values:
prior_H = 0.01  # 1% chance of having the disease
likelihood_D_given_H = 0.95  # 95% chance of a positive test if the person has the disease
likelihood_D_given_not_H = 0.05  # 5% false positive rate
prior_not_H = 0.99  # 99% chance of not having the disease

# Calculate the posterior probability using Bayes' Theorem
posterior_H_given_D = bayes_theorem(prior_H, likelihood_D_given_H, likelihood_D_given_not_H, prior_not_H)

print(f"The probability of having the disease given a positive test result is: {posterior_H_given_D:.4f}")
```

```python
# 12.Perform a Chi-square test for independence between two categorical variables in Python.
import numpy as np
import scipy.stats as stats
import pandas as pd

# Example categorical data (e.g., survey results)
# Create a 2x2 contingency table: Rows represent Gender (Male, Female), Columns represent Preference (Yes, No)
data = np.array([[30, 10],   # Male: 30 Yes, 10 No
                 [20, 20]])  # Female: 20 Yes, 20 No

# Create a DataFrame for better visualization
df = pd.DataFrame(data, columns=["Yes", "No"], index=["Male", "Female"])
print("Contingency Table:")
print(df)

# Perform the Chi-Square Test for Independence
chi2_stat, p_val, dof, expected = stats.chi2_contingency(data)

print("\nChi-Square Test Results:")
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

# Interpret the result
alpha = 0.05  # Significance level

if p_val < alpha:
    print("\nReject the null hypothesis: There is a significant association between the variables.")
else:
    print("\nFail to reject the null hypothesis: There is no significant association between the variables.")
```

```python
# 13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
# data.
import numpy as np

# Function to calculate expected frequencies for a Chi-Square test
def calculate_expected_frequencies(observed_data):
    # Get the total number of observations (N)
    total = np.sum(observed_data)

    # Get the row sums and column sums
    row_sums = np.sum(observed_data, axis=1)
    col_sums = np.sum(observed_data, axis=0)

    # Calculate the expected frequencies matrix
    expected_data = np.outer(row_sums, col_sums) / total

    return expected_data

# Example observed data (Contingency Table)
# Example: Survey results on two variables, e.g., Gender (Male/Female) vs Preference (Yes/No)
observed_data = np.array([[30, 10],  # Male: 30 Yes, 10 No
                          [20, 20]]) # Female: 20 Yes, 20 No

# Calculate expected frequencies
expected_data = calculate_expected_frequencies(observed_data)

print("Observed Frequencies:")
print(observed_data)

print("\nExpected Frequencies:")
print(expected_data)
```

```python
# 14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

import numpy as np
import scipy.stats as stats

# Observed data (frequency of each outcome in a 6-sided die roll)
observed = np.array([12, 8, 11, 7, 9, 13])

# Total number of observations (sum of all observed values)
total_rolls = np.sum(observed)

# Expected frequency (for a fair die, each outcome should occur with equal probability)
expected = np.array([total_rolls / 6] * 6)

# Perform the Chi-Square Goodness-of-Fit Test
chi2_stat, p_val = stats.chisquare(observed, expected)

# Output the test statistic and p-value
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_val:.4f}")

# Interpret the result
alpha = 0.05  # Significance level

if p_val < alpha:
    print("\nReject the null hypothesis: The observed data does not follow the expected distribution.")
else:
    print("\nFail to reject the null hypothesis: The observed data follows the expected distribution.")
```

```python
# 16.Implement an F-test using Python to compare the variances of two random samples
import numpy as np
import scipy.stats as stats

# Function to perform an F-test to compare the variances of two samples
def f_test(sample1, sample2):
    # Calculate the sample variances
    var1 = np.var(sample1, ddof=1)  # Sample variance for sample1
    var2 = np.var(sample2, ddof=1)  # Sample variance for sample2

    # Calculate the F-statistic (larger variance / smaller variance)
    f_statistic = var1 / var2 if var1 > var2 else var2 / var1

    # Degrees of freedom for each sample
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1

    # Calculate the p-value from the F-distribution
    p_value = 2 * min(stats.f.cdf(f_statistic, df1, df2), 1 - stats.f.cdf(f_statistic, df1, df2))

    return f_statistic, p_value

# Example: Generate two random samples with different variances
np.random.seed(42)
sample1 = np.random.normal(loc=0, scale=5, size=100)  # Sample 1 (mean=0, std=5)
sample2 = np.random.normal(loc=0, scale=3, size=100)  # Sample 2 (mean=0, std=3)

# Perform the F-test
f_stat, p_val = f_test(sample1, sample2)

# Output the results
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_val:.4f}")

# Interpret the result
alpha = 0.05  # Significance level

if p_val < alpha:
    print("\nReject the null hypothesis: The variances of the two samples are significantly different.")
else:
    print("\nFail to reject the null hypothesis: The variances of the two samples are not significantly different.")
```

```python
# 17.Write a Python program to perform an ANOVA test to compare means between multiple groups and
# interpret the results.
import numpy as np
import scipy.stats as stats

# Function to perform a one-way ANOVA test
def perform_anova(*samples):
    # Perform the ANOVA test using scipy's f_oneway function
    f_stat, p_value = stats.f_oneway(*samples)

    return f_stat, p_value

# Example data: Scores of students from three different teaching methods
np.random.seed(42)

# Group 1: Teaching Method 1
group1 = np.random.normal(loc=75, scale=10, size=30)

# Group 2: Teaching Method 2
group2 = np.random.normal(loc=80, scale=12, size=30)

# Group 3: Teaching Method 3
group3 = np.random.normal(loc=85, scale=15, size=30)

# Perform the ANOVA test
f_stat, p_value = perform_anova(group1, group2, group3)

# Output the results
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of the result
alpha = 0.05  # Significance level

if p_value < alpha:
    print("\nReject the null hypothesis: There is a significant difference between the group means.")
else:
    print("\nFail to reject the null hypothesis: There is no significant difference between the group means.")
```

```python
# 19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.
import numpy as np
import scipy.stats as stats

def check_anova_assumptions(groups):
    """
    Function to check the assumptions of normality, equal variances, and independence for one-way ANOVA.

    Parameters:
    groups (list of arrays): List of groups where each group is an array of data points.

    Returns:
    dict: Dictionary containing results for normality, equal variance, and independence assumptions.
    """
    results = {}

    # 1. Check Normality using Shapiro-Wilk test
    normality_results = []
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        normality_results.append((f'Group {i+1}', stat, p_value))

    results['normality'] = normality_results

    # 2. Check Equal Variance using Levene's test
    stat, p_value = stats.levene(*groups)
    results['equal_variance'] = (stat, p_value)

    # 3. Check Independence assumption (Typically assessed by study design, not statistical test)
    # Independence is not directly tested, but it's good practice to note this assumption.
    results['independence'] = 'Assumed based on random sampling or study design'

    return results

# Example usage with some sample data
np.random.seed(0)

# Simulating 3 groups of data
group_A = np.random.normal(loc=50, scale=10, size=30)
group_B = np.random.normal(loc=55, scale=10, size=30)
group_C = np.random.normal(loc=60, scale=10, size=30)

groups = [group_A, group_B, group_C]

# Check assumptions for the groups
assumption_results = check_anova_assumptions(groups)

# Print the results
print("Assumptions Check Results:")
for assumption, result in assumption_results.items():
    if assumption == 'normality':
        print("\nNormality Test Results:")
        for group_name, stat, p_val in result:
            print(f"{group_name} - Shapiro-Wilk Stat: {stat:.4f}, p-value: {p_val:.4f}")
    elif assumption == 'equal_variance':
        stat, p_val = result
        print(f"\nEqual Variance Test (Levene's Test): Stat: {stat:.4f}, p-value: {p_val:.4f}")
    else:
        print(f"\n{assumption}: {result}")
```

```python
# 20.Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the
# resultsD
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate sample data for two-way ANOVA
np.random.seed(0)  # For reproducibility

# Simulating data for two factors: Factor A (2 levels), Factor B (3 levels)
factor_A = ['A1', 'A2']
factor_B = ['B1', 'B2', 'B3']

# Creating a DataFrame with sample data
data = {
    'Factor_A': np.repeat(factor_A, 30),  # 30 samples per group of Factor A
    'Factor_B': np.tile(factor_B, 20),  # 20 samples per group of Factor B
    'Value': np.random.normal(loc=50, scale=10, size=60)  # Normally distributed dependent variable
}

df = pd.DataFrame(data)

# Introduce some interaction effect
df.loc[df['Factor_A'] == 'A1', 'Value'] += 5
df.loc[df['Factor_B'] == 'B2', 'Value'] -= 5

# Step 2: Perform Two-Way ANOVA using statsmodels
# Fit the model: 'Value' ~ 'Factor_A' + 'Factor_B' + 'Factor_A:Factor_B'
model = smf.ols('Value ~ C(Factor_A) + C(Factor_B) + C(Factor_A):C(Factor_B)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Print the ANOVA table
print("Two-Way ANOVA Results:")
print(anova_table)

# Step 3: Visualize the interaction effect using a plot
plt.figure(figsize=(8, 6))
sns.pointplot(x='Factor_B', y='Value', hue='Factor_A', data=df, dodge=True, markers='o', linestyles='-', ci=None)

# Customizing the plot
plt.title('Interaction Plot: Factor_A x Factor_B')
plt.xlabel('Factor B')
plt.ylabel('Values')
plt.show()
```

```python
# 23.Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Step 1: Simulate random data from a normal distribution
np.random.seed(0)  # For reproducibility

# Simulating data for two groups (Sample 1 and Sample 2)
sample_1 = np.random.normal(loc=50, scale=10, size=100)  # mean = 50, std dev = 10
sample_2 = np.random.normal(loc=55, scale=10, size=100)  # mean = 55, std dev = 10

# Step 2: Perform two-sample t-test
t_statistic, p_value = stats.ttest_ind(sample_1, sample_2)

# Step 3: Print the results of the t-test
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Step 4: Interpret the p-value
if p_value < 0.05:
    print("We reject the null hypothesis: The means are significantly different.")
else:
    print("We fail to reject the null hypothesis: The means are not significantly different.")

# Step 5: Visualize the data with boxplots to compare the means
plt.figure(figsize=(8, 6))
plt.boxplot([sample_1, sample_2], labels=["Sample 1", "Sample 2"])
plt.title('Comparison of Sample Means')
plt.ylabel('Values')
plt.show()
```

```python
# 24.
import numpy as np
import scipy.stats as stats

# Simulate data (e.g., sample size = 30, sample variance = 100, population variance = 90)
np.random.seed(0)
sample_data = np.random.normal(loc=50, scale=10, size=30)  # sample with a standard deviation of 10

# Sample statistics
n = len(sample_data)
sample_variance = np.var(sample_data, ddof=1)

# Hypothesized population variance
population_variance = 90

# Chi-square statistic
chi_square_stat = (n - 1) * sample_variance / population_variance

# Degrees of freedom
df = n - 1

# P-value
p_value = 2 * min(stats.chi2.cdf(chi_square_stat, df), 1 - stats.chi2.cdf(chi_square_stat, df))

# Print results
print(f"Chi-square Statistic: {chi_square_stat}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: The sample variance is significantly different from the population variance.")
else:
    print("Fail to reject the null hypothesis: The sample variance is not significantly different from the population variance.")
```

```python
# 25
import numpy as np
import scipy.stats as stats

# Simulate data for two groups (Group 1 and Group 2)
n1, n2 = 100, 120  # sample sizes
x1, x2 = 60, 75  # number of successes

# Sample proportions
p1 = x1 / n1
p2 = x2 / n2

# Pooled proportion
p = (x1 + x2) / (n1 + n2)

# Standard error of the difference in proportions
se = np.sqrt(p * (1 - p) * (1 / n1 + 1 / n2))

# Z-statistic
z_stat = (p1 - p2) / se

# P-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Print results
print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: The proportions are not significantly different.")
```
